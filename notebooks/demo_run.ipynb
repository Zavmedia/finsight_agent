## 2. Running the Agent

Make sure your FastAPI server is running. Then, you can call the `/agent` endpoint with your query.

```python
import requests
import json

API_URL = "http://localhost:8000"
user_query = "Provide a financial analysis for Microsoft (MSFT)."

print(f"Sending query: {user_query}")

agent_response = requests.post(f"{API_URL}/agent", json={"query": user_query})

if agent_response.status_code == 200:
    print("Agent run successful!")
    final_state = agent_response.json().get('final_state', {})
    print("\n--- Final Report ---\n")
    print(final_state.get('report', 'No report generated.'))
else:
    print(f"Agent run failed with status code: {agent_response.status_code}")
    print(agent_response.text)
```
